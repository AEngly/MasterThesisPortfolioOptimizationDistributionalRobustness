# Testing Affine Functions for Corollary 5.1 (DRO)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def first(x):
    return np.maximum(x, -x);

def second(x, tau):
    return np.maximum(-x - tau, 0);

def actualLossFunction(x,tau,gamma):
    return first(x) + gamma*tau + (gamma/(1-beta))*second(x, tau)

def affine1(x, tau, gamma):
    ak = (1 - gamma/(1 - beta));
    bk = (gamma*tau*(1 - 1/(1-beta)));
    return(ak*x + bk);

def affine2(x, tau, gamma):
    ak = -(1 + gamma/(1 - beta));
    bk = (gamma*tau*(1 - 1/(1-beta)));
    return(ak*x + bk);

def affine3(x, tau, gamma):
    ak = 1;
    bk = gamma*tau;
    return(ak*x + bk);

def affine4(x, tau, gamma):
    ak = -1;
    bk = gamma*tau;
    return(ak*x + bk);

def affineConstruction(x, tau, gamma):
    return np.maximum(np.maximum(affine1(x, tau, gamma), affine2(x, tau, gamma)), np.maximum(affine3(x, tau, gamma), affine4(x, tau, gamma)))
```

```python
x = np.linspace(-1,1,1000);

tau = 0.05;
beta = 0.80;
gamma = 0.8;

plt.figure(figsize=(16,10));
#plt.plot(x, actualLossFunction(x,tau,gamma), linewidth=6, color = "lightblue", label=r"$[True] \ \ \tau={}$".format(tau));
#plt.plot(x, affine1(x, tau, gamma), label=r"$[1] \ \ \tau={}$".format(tau));
#plt.plot(x, affine2(x, tau, gamma), label=r"$[2] \ \ \tau={}$".format(tau));
#plt.plot(x, affine3(x, tau, gamma), label=r"$[3] \ \ \tau={}$".format(tau));
#plt.plot(x, affine4(x, tau, gamma), label=r"$[4] \ \ \tau={}$".format(tau));
#plt.plot(x, affineConstruction(x, tau, gamma), label=r"$[Max] \ \ \tau={}$".format(tau), linewidth=6, linestyle='--', color="red");
plt.plot(x, affineConstruction(x, tau, gamma)-actualLossFunction(x,tau,gamma), label=r"$[4] \ \ \tau={}$".format(tau));
plt.legend();
plt.show();
```
