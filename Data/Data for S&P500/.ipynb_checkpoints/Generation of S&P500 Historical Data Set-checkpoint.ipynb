```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Load constituents and there tickers.

```python
TICKERS = [
    "AMZN", # Amazon (CONSUMER DISCRETIONARY)
    "V", # Visa (FINANCIALS)
    "BRK-B", # Birkshire Hathaway (FINANCIALS)
    "BMY", # Bristol-Meyers Squibb Company (HEALTH CARE)
    "QCOM", # Qualcomm (INFORMATION TECHNOLOGY)
    "CSCO", # Cisco (INFORMATION TECHNOLOGY)
    "KO", # Coca Cola (CONSUMER STABLES)
    "NEE" #NextEra Energi Inc. (ENERGY)
]
```

```python
constituents = pd.read_csv(r'S&P 500 Historical Components & Changes.csv')
```

```python
TICKERS = constituents['tickers'].iloc[-1].split(",")
```

Fetch the data from Yahoo Finance.

```python
START = "2016-01-01"
END = "2022-01-01"

INTERVAL = "1mo"
stock_data = yf.download(
    tickers = TICKERS[:10],
    start = START,
    end = END,
    interval = INTERVAL
).dropna()['Adj Close']
```
