```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate random samples from two normal distributions
mean1, mean2 = 0, 0
std_dev1, std_dev2 = 1, 2
size = 1000

x_samples = np.random.normal(mean1, std_dev1, size)
y_samples = np.random.normal(mean2, std_dev2, size)

# Create a joint plot with marginal distributions using seaborn
sns.set(style="whitegrid")
g = sns.jointplot(x=x_samples, y=y_samples, kind="scatter", height=7)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['text.usetex'] = True
plt.rcParams['mathtext.fontset'] = 'custom'
plt.rcParams['mathtext.rm'] = 'Bitstream Vera Sans'
plt.rcParams['mathtext.it'] = 'Bitstream Vera Sans:italic'
plt.rcParams['mathtext.bf'] = 'Bitstream Vera Sans:bold'
def_font_size = 12
plt.rcParams.update({'font.size': def_font_size})

# Colors
myMap = plt.get_cmap('inferno')  # Paired, inferno
colors = myMap(np.linspace(0, 1, 10))

# Set random seed for reproducibility
np.random.seed(42)

# Generate random samples with correlation
mean1, mean2 = 0, 0
std_dev1, std_dev2 = 1, 2
size = 10000
correlation_coefficient = 0.8

# Create covariance matrix
covariance_matrix = np.array([[std_dev1**2, correlation_coefficient * std_dev1 * std_dev2],
                              [correlation_coefficient * std_dev1 * std_dev2, std_dev2**2]])

# Generate correlated samples
correlated_samples = np.random.multivariate_normal([mean1, mean2], covariance_matrix, size)

# Extract correlated x and y samples
x_samples = correlated_samples[:, 0]
y_samples = correlated_samples[:, 1]

# Create a joint plot with marginal distributions
sns.set(style="white", color_codes=True)
g = sns.jointplot(x=x_samples, y=y_samples, kind="scatter", height=7, color=colors[0])

# Customize marginal plots
g.ax_marg_x.hist(x_samples, color=colors[4], bins=np.arange(-5, 5, 0.5), alpha=0.7)
g.ax_marg_y.hist(y_samples, color=colors[6], bins=np.arange(-10, 10, 1), orientation="horizontal", alpha=0.7)

# Show the plot
plt.show()
```

```python
import numpy as np
from scipy.stats import wasserstein_distance
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
# np.random.seed(42)

# Generate random samples with correlation
mean1, mean2 = 0, 0
std_dev1, std_dev2 = 1, 2
size = 1000
correlation_coefficient = 0.8

# Create covariance matrix
covariance_matrix = np.array([[std_dev1**2, correlation_coefficient * std_dev1 * std_dev2],
                              [correlation_coefficient * std_dev1 * std_dev2, std_dev2**2]])

# Generate correlated samples
correlated_samples = np.random.multivariate_normal([mean1, mean2], covariance_matrix, size)

# Extract correlated x and y samples
x_samples = correlated_samples[:, 0]
y_samples = correlated_samples[:, 1]

# Create histograms for marginal distributions
hist_x, bins_x = np.histogram(x_samples, bins=50, density=True)
hist_y, bins_y = np.histogram(y_samples, bins=50, density=True)

# Calculate the Wasserstein distance using the 1-norm
wasserstein_distance_1norm = wasserstein_distance(hist_x, hist_y)

# Print the calculated Wasserstein distance
print(f"Wasserstein Distance (1-norm): {wasserstein_distance_1norm}")

# Plot histograms for visualization
sns.histplot(x_samples, bins=bins_x, kde=True, color='skyblue', label='X')
sns.histplot(y_samples, bins=bins_y, kde=True, color='salmon', label='Y')
plt.legend()
plt.title('Marginal Distributions')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```
