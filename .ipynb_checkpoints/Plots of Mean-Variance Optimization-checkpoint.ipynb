## Efficient Frontier and Capital Market Line: Analytical

We start by showing the efficient frontier via the analytical expression obtained in the thesis.

```python
def OptimalPortfolio(targetReturn, mu, Sigma, riskFree=False, rf=0.05):
    
    if not riskFree:
    
        e = np.ones(len(mu));
        aux1 = np.vstack((mu,e)).transpose();
        aux2 = np.array([targetReturn, 1]);
        invSigma = np.linalg.inv(Sigma);
        A = np.dot(np.dot(aux1.transpose(), invSigma), aux1);
        invA = np.linalg.inv(A);

        # Then we find the optimal portfolio
        x_star = np.dot(invSigma, np.dot(aux1, np.dot(invA, aux2)));
        mu_star = np.dot(x_star,mu);
        sigma_star = np.sqrt(np.dot(aux2, np.dot(invA, aux2)));

        # In addition, we compute the global min variance portfolio.
        gmv_star = 1/A[1,1] * np.dot(invSigma, e);
        sigma_gmv = np.sqrt(np.dot(gmv_star, np.dot(Sigma, gmv_star)));
        mu_gmv = np.dot(gmv_star,mu);

        return (x_star, mu_star, sigma_star, mu_gmv, sigma_gmv);
    
    else:
        
        # Create auxiliaries
        mu_e = mu - rf;
        e = np.ones(len(mu));
        
        # Calculate capital market line
        invSigma = np.linalg.inv(Sigma);
        factor = np.sqrt(mu_e.T@invSigma@mu_e)
        sigma_star = (1/factor)*targetReturn
        mu_star = targetReturn
        
        # Calculate tangency portfolio
        mu_tangency = factor * 1/(e.T@invSigma@mu_e)
        sigma_tangency = (1/factor)*(mu_tangency-rf)

        return (mu_star, sigma_star, mu_tangency, sigma_tangency);
        
        
```

Then we define a sequence of target returns for which we want to find the optimal portfolio.

```python
targetReturns = np.linspace(0.000, 0.50, num=500);
rf = 0.03
excessReturns = np.linspace(0.000, 0.50-rf, num=500);
sigma = np.sqrt(np.diag(Sigma));

# Efficient Frontier (EF)
optimalReturns = [];
optimalStandardDeviations = [];
optimalPortfolios = [];

for targetReturn in targetReturns:
    
    x_star, mu_star, sigma_star, mu_gmv, sigma_gmv = OptimalPortfolio(targetReturn, mu, Sigma);
    optimalPortfolios.append(x_star);
    optimalReturns.append(mu_star);
    optimalStandardDeviations.append(sigma_star);

optimalReturns = np.array(optimalReturns);
optimalStandardDeviations = np.array(optimalStandardDeviations); 

# Capital Market Line (CML)
optimalReturnsCML = [];
optimalStandardDeviationsCML = [];

for excessReturn in excessReturns:
    
    mu_star, sigma_star, mu_tangency, sigma_tangency = OptimalPortfolio(excessReturn, mu, Sigma, riskFree=True, rf=rf);
    optimalReturnsCML.append(mu_star);
    optimalStandardDeviationsCML.append(sigma_star);

optimalReturnsCML = np.array(optimalReturnsCML)+rf;
optimalStandardDeviationsCML = np.array(optimalStandardDeviationsCML); 
```

## Efficient Frontier: Linear Program with Conic Constraint

We follow the implementation from MOSEK's cookbook.

```python
N = mu.shape[0] # Number of securities 
gamma2 = 0.05 # Risk limit

# Cholesky factor of S to use in conic risk constraint
G = np.linalg.cholesky(Sigma)
```

```python
def MVO(CF, MV, N, shorting=True, gammas = np.linspace(0.02, 0.115, 10000)):
    
    G = CF;
    mu = MV;

    with Model("markowitz") as M: 

        # Settings
        #M.setLogHandler(sys.stdout)

        # Decision variable (fraction of holdings in each security)
        # The variable x is the fraction of holdings in each security.
        if shorting:
            x = M.variable("x", N, Domain.unbounded())
        else:
            x = M.variable("x", N, Domain.greaterThan(0.0))

        # Budget constraint
        M.constraint('budget', Expr.sum(x), Domain.equalsTo(1)) 

        # Objective
        gamma2 = M.parameter('gamma2');
        M.objective('obj', ObjectiveSense.Maximize, Expr.dot(mu, x))

        # Conic constraint for the portfolio variance
        M.constraint('risk', Expr.vstack(gamma2, 0.5, Expr.mul(G.T, x)), Domain.inRotatedQCone())

        columns = ["delta", "obj", "return", "risk"] + ["PM", "LMT", "MCD", "MMM", "AAPL", "MSFT", "TXN", "CSCO"]
        df_result = pd.DataFrame(columns=columns)

        for idx, gamma in enumerate(gammas):

            # Update parameter
            gamma2.setValue(gamma);

            # Solve optimization
            M.solve();

            prosta = M.getProblemStatus();

            if prosta !=ProblemStatus.PrimalInfeasible:
                
                # Print status
                # print("The program is feasible for gamma = {}.".format(gamma));
                
                # Save results
                portfolio_return = mu @ x.level();
                portfolio_risk = np.sqrt(x.level().T @ Sigma @ x.level());
                row = pd.DataFrame([gamma, M.primalObjValue(), portfolio_return,
                                 portfolio_risk] + list(x.level()), index=columns, columns=[idx]);

                # Concatenate with exisiting results
                df_result = pd.concat([df_result, row.T], axis=0);

            else:
                continue;
                # print("The program is primal infeasible for gamma = {}.".format(gamma));
                
        return df_result;
    
def MVO_Cardinality(CF, MV, N, x0, K=4, deltas = np.linspace(0.02, 0.115, 100), shorting=True):
    
    G = CF;
    m = MV;
    
    # Set big-M constant
    xp_ub = 1*np.ones(N)
    xm_ub = 1*np.ones(N)
    x_ub = 1*np.ones(N)

    with Model("Case study") as M:
        
        # --- Real variables ---
        
        # The variable x is the fraction of holdings in each security.
        if shorting:
            x = M.variable("x", N, Domain.unbounded())
        else:
            x = M.variable("x", N, Domain.greaterThan(0.0))
            
        # The variable xp is the fraction of the wealth that is spend to purchase from x
        xp = M.variable("xp", N, Domain.greaterThan(0.0))
        # The variable xm is the fraction of the wealth that is freed up by selling from x
        xm = M.variable("xm", N, Domain.greaterThan(0.0))

        # --- Binary variables ---
        
        # The binary variable yp denotes whether shares are bought
        yp = M.variable("yp", N, Domain.binary())
        # The binary variable ym denotes whether shares are sold
        ym = M.variable("ym", N, Domain.binary())
        # The binary variable y denotes whether shares of i are owned
        y = M.variable("y", N, Domain.binary())
        
        # --- Constraints ---

        # Traded amount: Constraint assigning xp and xm to the pos. and neg. part of x.
        M.constraint('pos-neg-part', Expr.sub(x, Expr.sub(xp, xm)),
                   Domain.equalsTo(0.0))
        
        # Make sure the binary i is turned on if asset i is bought
        M.constraint('control-positive-binaries', Expr.sub(xp, Expr.mulElm(yp, xp_ub)),
                   Domain.lessThan(0.0))
        M.constraint('control-negative-binaries', Expr.sub(xm, Expr.mulElm(ym, xm_ub)),
                   Domain.lessThan(0.0))
        M.constraint('record-ownership', Expr.sub(x, Expr.mulElm(y, x_ub)),
                   Domain.lessThan(0.0))
        
        # Add cardinality constraint
        M.constraint('cardinality constraint', Expr.sum(y),
                   Domain.lessThan(K))        

        # Exclusive buy-sell constraint
        M.constraint('exclusion', Expr.add(yp, ym), Domain.lessThan(1.0))

        # Budget constraint
        M.constraint('budget', Expr.sum(x), Domain.equalsTo(1.0))

        # Conic constraint for the portfolio variance
        delta = M.parameter()
        M.constraint('risk', Expr.vstack(delta, 0.5, Expr.mul(G.transpose(), x)),
                   Domain.inRotatedQCone())
        
        # --- Objective function ---
        
        # Objective function
        M.objective('obj', ObjectiveSense.Maximize, Expr.dot(m, x))
        
        # --- Solve model and record results ---

        assets = ["PM", "LMT", "MCD", "MMM", "AAPL", "MSFT", "TXN", "CSCO"]
        columns = ["delta", "obj", "return", "risk", "cardinality", "x_sum"] + assets;
        df_result = pd.DataFrame(columns=columns)
        
        for idx, d in enumerate(deltas):
            
            # Update parameter
            delta.setValue(d)

            # Solve optimization
            M.solve()
            
            # Get problem status
            prosta = M.getProblemStatus();
            
            if prosta != ProblemStatus.PrimalInfeasible:
                
                # Get the weights
                xbar = x.level()
                portfolio_return = m @ xbar  
                portfolio_risk = np.sqrt(xbar.T @ Sigma @ xbar);
                
                # Find cardinality
                cardinality = sum(x.level() > 0)
                
                # Collect results
                row = pd.DataFrame([d, M.primalObjValue(), portfolio_return,
                                 portfolio_risk, cardinality, sum(xbar)] + list(xbar), index=columns, columns=[idx]);

                # Concatenate with exisiting results
                df_result = pd.concat([df_result, row.T], axis=0);
                
            else:
                continue;
                # print("The program is primal infeasible for delta = {}.".format(d));

        return df_result;
```

```python
searchSpace = np.linspace(0.013, 0.115, 1000)
df_result_shorting = MVO(G, mu, N, gammas=searchSpace, shorting=True)
df_result = MVO(G, mu, N, gammas=searchSpace, shorting=False)
df_result_cardinality = MVO_Cardinality(G, mu, N, np.zeros(len(mu)), K=3, deltas=searchSpace, shorting=False);
df_result_cardinality_shorting = MVO_Cardinality(G, mu, N, np.zeros(len(mu)), K=3, deltas=searchSpace, shorting=True);
```

```python
plt.figure(figsize=(12,12));

# Layout configurations
plt.rcParams['xtick.labelsize'] = def_font_size;
plt.rcParams['ytick.labelsize'] = def_font_size;

# Plot assets
plt.scatter(sigma, mu, c=POINT_COLORS, s=200);

# Get returns and risk
optimalReturnsConic = df_result['return'].values;
optimalStandardDeviationsConic = df_result['risk'].values;

optimalReturnsConicShorting = df_result_shorting['return'].values;
optimalStandardDeviationsConicShorting = df_result_shorting['risk'].values;

optimalReturnsConicCardinality = df_result_cardinality['return'].values;
optimalStandardDeviationsConicCardinality = df_result_cardinality['risk'].values;

optimalReturnsConicCardinalityShorting = df_result_cardinality_shorting['return'].values;
optimalStandardDeviationsConicCardinalityShorting = df_result_cardinality_shorting['risk'].values;

# Plot above global mean-variance portfolio
selectedRange = optimalReturns > mu_gmv;
plt.plot(optimalStandardDeviations[selectedRange == False], optimalReturns[selectedRange == False], c="red");
plt.plot(optimalStandardDeviationsConicShorting, optimalReturnsConicShorting, c='black', label="Efficient Frontier (Conic Program)");
plt.plot(optimalStandardDeviationsConicCardinalityShorting, optimalReturnsConicCardinalityShorting, c=POINT_COLORS[4], label=r"Efficient Frontier with $K = 3$ (Conic Program)");
plt.plot(optimalStandardDeviationsConic, optimalReturnsConic, c='black', linewidth=2, linestyle="dotted", label="Efficient Frontier without Shorting (Conic Program)");
plt.plot(optimalStandardDeviationsConicCardinality, optimalReturnsConicCardinality, c=POINT_COLORS[3], label=r"Efficient Frontier without Shorting and $K = 3$ (Conic Program)");


# Plot global minimum-variance portfolio
plt.scatter(sigma_gmv, mu_gmv, color='red', marker="^", s=200, zorder=2, label="Global Minimum-Variance Portfolio");

# Plot tangency portfolio
plt.scatter(sigma_tangency, mu_tangency, marker="^", color=COLORS[0], s=200, zorder=2, label="Tangency Portfolio");

# Show legend and axis labels
plt.xlabel("Risk ($\sigma$)", fontsize=def_font_size+2);
plt.ylabel("Expected Return ($\mu$)", fontsize=def_font_size+2);
plt.legend(fontsize=def_font_size+2, loc="upper left");
plt.xlim((0,0.4))
plt.ylim((-0.05,0.5))

# Save and show plot
plt.savefig('./PlotsTheory/MVO/Chapter2_BT_EfficientFrontierConstraints.pdf', dpi=150);
plt.show()
```

## Efficient Frontier: Extension with Transaction Costs

Like on the MOSEK tutorial we classify transaction costs into the following:

1. Fixed costs are independent of transaction volume. These include brokerage commissions and transfer fees.
2. Variable costs depend on the transaction volume. These comprise execution costs such as market impact, bid/ask spread, or slippage; and opportunity costs of failed or incomplete execution.

We implement the model in Equation (6.1).

```python
def MVO_TC(CF, MV, N, x0, deltas = np.linspace(0.02, 0.115, 100), controller=1, shorting=True, normalized=True):
    
    G = CF;
    m = MV;
    
    # Set fixed transaction costs
    fp = 10 * np.ones(N) * controller * (10**(-4)); # Depends on portfolio value
    fm = 20 * np.ones(N) * controller * (10**(-4)); # Depends on portfolio value
    
    # Set variable transaction costs
    vp = 70 * np.ones(N) * controller * (10**(-4));
    vm = 60 * np.ones(N) * controller * (10**(-4));
    
    # Set big-M constant
    xp_ub = 1*np.ones(N)
    xm_ub = 1*np.ones(N)

    with Model("Case study") as M:
        
        # --- Real variables ---
        
        # The variable x is the fraction of holdings in each security.
        if shorting:
            x = M.variable("x", N, Domain.unbounded())
        else:
            x = M.variable("x", N, Domain.greaterThan(0.0))
        # The variable xp is the fraction of the wealth that is spend to purchase from x
        xp = M.variable("xp", N, Domain.greaterThan(0.0))
        # The variable xm is the fraction of the wealth that is freed up by selling from x
        xm = M.variable("xm", N, Domain.greaterThan(0.0))

        # --- Binary variables ---
        
        # The binary variable yp denotes whether shares are bought
        yp = M.variable("yp", N, Domain.binary())
        # The binary variable ym denotes whether shares are sold
        ym = M.variable("ym", N, Domain.binary())

        # Traded amount: Constraint assigning xp and xm to the pos. and neg. part of x.
        M.constraint('pos-neg-part', Expr.sub(x, Expr.sub(xp, xm)),
                   Domain.equalsTo(0.0))
        
        # Make sure the binary i is turned on if asset i is bought
        M.constraint('control-positive-binaries', Expr.sub(xp, Expr.mulElm(yp, xp_ub)),
                   Domain.lessThan(0.0))
        M.constraint('control-negative-binaries', Expr.sub(xm, Expr.mulElm(ym, xm_ub)),
                   Domain.lessThan(0.0))

        # Exclusive buy-sell constraint
        M.constraint('exclusion', Expr.add(yp, ym), Domain.lessThan(1.0))

        # Budget constraint with transaction cost terms
        fixcost_terms = Expr.add([Expr.dot(fp, yp), Expr.dot(fm, ym)])
        varcost_terms = Expr.add([Expr.dot(vp, xp), Expr.dot(vm, xm)])
        transcost_terms = Expr.add([varcost_terms, fixcost_terms])
        budget_terms = Expr.add([Expr.sum(x), varcost_terms, fixcost_terms])
        M.constraint('budget', budget_terms, Domain.equalsTo(1.0))

        # Objective function
        delta = M.parameter()
        M.objective('obj', ObjectiveSense.Maximize, Expr.dot(m, x))

        # Conic constraint for the portfolio variance
        M.constraint('risk', Expr.vstack(delta, 0.5, Expr.mul(G.transpose(), x)),
                   Domain.inRotatedQCone())

        assets = ["PM", "LMT", "MCD", "MMM", "AAPL", "MSFT", "TXN", "CSCO"]
        columns = ["delta", "obj", "return", "risk", "x_sum", "tcost"] + assets;
        df_result = pd.DataFrame(columns=columns)
        
        for idx, d in enumerate(deltas):
            
            # Update parameter
            delta.setValue(d)

            # Solve optimization
            M.solve()
            
            # Get problem status
            prosta = M.getProblemStatus();
            
            if prosta != ProblemStatus.PrimalInfeasible:
                
                # Calculate transaction costs
                tcost = np.dot(vp, xp.level()) + np.dot(vm, xm.level()) + np.dot(fp, yp.level()) + np.dot(fm, ym.level())

                # Then we calculate the appropriate metrics
                if normalized:
                    xbar = x.level()/sum(x.level())
                    portfolio_return = m @ xbar - tcost
                else:
                    xbar = x.level()
                    portfolio_return = m @ xbar
                    
                portfolio_risk = np.sqrt(xbar.T @ Sigma @ xbar);
                
                # Collect results
                row = pd.DataFrame([d, M.primalObjValue(), portfolio_return,
                                 portfolio_risk, sum(xbar), tcost] + list(xbar), index=columns, columns=[idx]);

                # Concatenate with exisiting results
                df_result = pd.concat([df_result, row.T], axis=0);
                
            else:
                continue;
                # print("The program is primal infeasible for delta = {}.".format(d));

        return df_result;
```

## Efficient Frontier: Extension with Cardinality Constraints

```python
def MVO_Cardinality(CF, MV, N, x0, K=4, deltas = np.linspace(0.02, 0.115, 100), shorting=True):
    
    G = CF;
    m = MV;
    
    # Set big-M constant
    xp_ub = 1*np.ones(N)
    xm_ub = 1*np.ones(N)
    x_ub = 1*np.ones(N)

    with Model("Case study") as M:
        
        # --- Real variables ---
        
        # The variable x is the fraction of holdings in each security.
        if shorting:
            x = M.variable("x", N, Domain.unbounded())
        else:
            x = M.variable("x", N, Domain.greaterThan(0.0))
            
        # The variable xp is the fraction of the wealth that is spend to purchase from x
        xp = M.variable("xp", N, Domain.greaterThan(0.0))
        # The variable xm is the fraction of the wealth that is freed up by selling from x
        xm = M.variable("xm", N, Domain.greaterThan(0.0))

        # --- Binary variables ---
        
        # The binary variable yp denotes whether shares are bought
        yp = M.variable("yp", N, Domain.binary())
        # The binary variable ym denotes whether shares are sold
        ym = M.variable("ym", N, Domain.binary())
        # The binary variable y denotes whether shares of i are owned
        y = M.variable("y", N, Domain.binary())
        
        # --- Constraints ---

        # Traded amount: Constraint assigning xp and xm to the pos. and neg. part of x.
        M.constraint('pos-neg-part', Expr.sub(x, Expr.sub(xp, xm)),
                   Domain.equalsTo(0.0))
        
        # Make sure the binary i is turned on if asset i is bought
        M.constraint('control-positive-binaries', Expr.sub(xp, Expr.mulElm(yp, xp_ub)),
                   Domain.lessThan(0.0))
        M.constraint('control-negative-binaries', Expr.sub(xm, Expr.mulElm(ym, xm_ub)),
                   Domain.lessThan(0.0))
        M.constraint('record-ownership', Expr.sub(x, Expr.mulElm(y, x_ub)),
                   Domain.lessThan(0.0))
        
        # Add cardinality constraint
        M.constraint('cardinality constraint', Expr.sum(y),
                   Domain.lessThan(K))        

        # Exclusive buy-sell constraint
        M.constraint('exclusion', Expr.add(yp, ym), Domain.lessThan(1.0))

        # Budget constraint
        M.constraint('budget', Expr.sum(x), Domain.equalsTo(1.0))

        # Conic constraint for the portfolio variance
        delta = M.parameter()
        M.constraint('risk', Expr.vstack(delta, 0.5, Expr.mul(G.transpose(), x)),
                   Domain.inRotatedQCone())
        
        # --- Objective function ---
        
        # Objective function
        M.objective('obj', ObjectiveSense.Maximize, Expr.dot(m, x))
        
        # --- Solve model and record results ---

        assets = ["PM", "LMT", "MCD", "MMM", "AAPL", "MSFT", "TXN", "CSCO"]
        columns = ["delta", "obj", "return", "risk", "cardinality", "x_sum"] + assets;
        df_result = pd.DataFrame(columns=columns)
        
        for idx, d in enumerate(deltas):
            
            # Update parameter
            delta.setValue(d)

            # Solve optimization
            M.solve()
            
            # Get problem status
            prosta = M.getProblemStatus();
            
            if prosta != ProblemStatus.PrimalInfeasible:
                
                # Get the weights
                xbar = x.level()
                portfolio_return = m @ xbar  
                portfolio_risk = np.sqrt(xbar.T @ Sigma @ xbar);
                
                # Find cardinality
                cardinality = sum(x.level() > 0)
                
                # Collect results
                row = pd.DataFrame([d, M.primalObjValue(), portfolio_return,
                                 portfolio_risk, cardinality, sum(xbar)] + list(xbar), index=columns, columns=[idx]);

                # Concatenate with exisiting results
                df_result = pd.concat([df_result, row.T], axis=0);
                
            else:
                continue;
                # print("The program is primal infeasible for delta = {}.".format(d));

        return df_result;
```

```python
df_result = MVO_noShorting(G, mu, N, gammas = deltas);
df_result_cardinality_2 = MVO_Cardinality(G, mu, N, x0, K=2, deltas = deltas, controller=1, shorting=False, normalized=True);
df_result_cardinality_3 = MVO_Cardinality(G, mu, N, x0, K=3, deltas = deltas, controller=1, shorting=False, normalized=True);
df_result_cardinality_4 = MVO_Cardinality(G, mu, N, x0, K=4, deltas = deltas, controller=1, shorting=False, normalized=True);
df_result_cardinality_5 = MVO_Cardinality(G, mu, N, x0, K=5, deltas = deltas, controller=1, shorting=False, normalized=True);
df_result_cardinality_6 = MVO_Cardinality(G, mu, N, x0, K=6, deltas = deltas, controller=1, shorting=False, normalized=True);
df_result_cardinality_7 = MVO_Cardinality(G, mu, N, x0, K=7, deltas = deltas, controller=1, shorting=False, normalized=True);
df_result_cardinality_8 = MVO_Cardinality(G, mu, N, x0, K=8, deltas = deltas, controller=1, shorting=False, normalized=True);
```

In the case of cardinality constraints, it clearly only affects the portfolio, if we want the risk to be low. This is because there are more assets to choose between in that case.
