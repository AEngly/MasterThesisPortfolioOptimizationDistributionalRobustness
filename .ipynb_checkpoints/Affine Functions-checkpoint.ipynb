# Testing Affine Functions for Corollary 5.1 (DRO)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def first(x):
    return np.maximum(x, -x);

def second(x, tau):
    return np.maximum(-x - tau, 0);

def affine1(x, tau):
    ak = (1 - gamma/(1 - beta));
    bk = (gamma*tau*(1 - 1/(1-beta)));
    return(ak*x + bk);

def affine2(x, tau):
    ak = -(1 + gamma/(1 - beta));
    bk = (gamma*tau*(1 - 1/(1-beta)));
    return(ak*x + bk);

def affine3(x, tau):
    ak = 1;
    bk = gamma*tau;
    return(ak*x + bk);

def affine4(x, tau):
    ak = -1;
    bk = gamma*tau;
    return(ak*x + bk);
```
