In this notebook, I make plots to support the understanding of the theory section.

```python
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt
```

## Figure 1: Value-at-Risk (VaR)

Suppose we have some distribution of financial returns with the following characteristics.

```python
mu = 0.09
sd = 0.15
```

Then we can plot the distribution.

```python
# Construct curve
x = np.linspace(-1, 1, 1000)
y_pdf = ss.norm.pdf(x, loc=mu, scale=sd)
y_cdf = ss.norm.cdf(x, loc=mu, scale=sd)
x_shade_pdf = np.linspace(min(x),ss.norm.ppf(0.20, loc=mu, scale=sd),100)
y_shade_pdf = ss.norm.pdf(x_shade, loc=mu, scale=sd)
x_shade_cdf = np.linspace(min(x),ss.norm.ppf(0.20, loc=mu, scale=sd),100)
y_shade_cdf = ss.norm.pdf(x_shade, loc=mu, scale=sd)

# Plot left side
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(16, 8))
axs[0].plot(x, y_pdf, alpha=0.5)
axs[0].fill_between(x_shade_pdf, y_shade_pdf, color='blue')

# Plot right side
axs[1].plot(x, y_cdf, alpha=0.5)
axs[1].plot(x, y_cdf, alpha=0.5)

plt.show()
```
