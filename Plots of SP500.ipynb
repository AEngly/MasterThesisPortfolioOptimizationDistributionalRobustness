Start by loading in the package.

```python
from EITP.DataHandlers.DataLoader import DataLoader;
import datetime as dt
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Then get the data.

```python
# Start by instantiating the data loader
dataLoader = DataLoader(path='./Data/');
allData = dataLoader.AggregateData(intersect=False, filtered=True, startDate="1999-01-01", endDate=dt.datetime.today().strftime("%Y-%m-%d"));
totalObservations = allData.shape[0]
```

```python
allData['SPX-INDEX'].values
```

```python
# Configure matplotlib to use LaTeX for rendering
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
plt.rc('font', size=18)

# Specify colors
myMap = plt.get_cmap('inferno') # Paired, inferno
colors = myMap(np.linspace(0.1,0.8,7))

fig, ax = plt.subplots(figsize=(12,8))
ax.plot(pd.to_datetime(allData['Dates']), 100*(allData['SPX-INDEX'].values/allData['SPX-INDEX'].values[0] - 1), label=r"S\&P 500 Float-Adjusted Market Capitalization Weighted Index", alpha=1, color=colors[5])
ax.set_xlabel('Years')
ax.set_ylabel('Cumulative Returns [\%]')
plt.legend()
plt.savefig('PlotsData/SP500.pdf', dpi=150)
plt.show()
```
