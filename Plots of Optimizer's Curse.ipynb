Define a simulation experiment.

```python
import matplotlib.pyplot as plt
import numpy as np

# Total assets
N = 10

# Set seed
np.random.seed(2345)

# Specify model parameters
mu = np.array([0.03 * i for i in range(1, N+1)])
cov = np.diag([(0.02 * i) ** 2 for i in range(1, N+1)])

# Create samples
samples = np.random.multivariate_normal(mu, cov, size=3, check_valid='warn', tol=1e-8)

# Calculate means and standard deviations
means = np.mean(samples, axis=0)
std_devs = np.std(samples, axis=0)

# Sample data
categories = [str(i) for i in range(1, N+1)]

# Get color schema
myMap = plt.get_cmap('inferno')  # Paired, inferno
colors = myMap(np.linspace(0, 1, 10))

# Create plot
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(12, 8))

# Create bar plot with error bars
plt.bar(categories, mu*100, capsize=5, color=colors[0], edgecolor='black', alpha=0.7, label="True Means")
plt.bar(categories, means*100, capsize=5, color=colors[5], edgecolor='black', alpha=0.7, label="Estimated Means")


# Add labels and title
plt.xlabel('Projects')
plt.ylabel('Return on Investments [\%]')
plt.legend()

# Save figure
plt.savefig('./PlotsTheory/OptimizersCurse1.pdf', dpi=150)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Select larges K
K = 4

# Set seed
np.random.seed(2345)

# Specify model parameters
mu = np.array([0.03 * i for i in range(1, N+1)])
cov = np.diag([(0.02 * i) ** 2 for i in range(1, N+1)])

# Create samples
samples = np.random.multivariate_normal(mu, cov, size=3, check_valid='warn', tol=1e-8)

# Calculate means and standard deviations
means = np.mean(samples, axis=0)
std_devs = np.std(samples, axis=0)

# Sample data
categories = [str(i) for i in range(1, N+1)]

# Sort categories and corresponding means
sorted_indices = np.argsort(means)
sorted_categories = np.array(categories)[sorted_indices]
sorted_means = means[sorted_indices]
sorted_std_devs = std_devs[sorted_indices]

# Get color schema
myMap = plt.get_cmap('inferno')  # Paired, inferno
colors = myMap(np.linspace(0, 1, 10))

# Create plot
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(12, 8))

# Create bar plot with error bars
plt.bar(sorted_categories, mu[sorted_indices]*100, capsize=5, color=colors[0], edgecolor='black', alpha=0.7, label="True Means")
plt.bar(sorted_categories, sorted_means*100, capsize=5, color=colors[5], edgecolor='black', alpha=0.7, label="Estimated Means")
plt.axhline(100*np.mean(mu[-K:]), label="Optimal Project Portfolio", linewidth=3, color=colors[0])
plt.axhline(100*np.mean(means[-K:]), label="In-Sample Performance of Chosen Portfolio", linestyle="--", linewidth=3, color=colors[5])
plt.axhline(100*np.mean(mu[sorted_indices][-K:]), label="Out-of-Sample Performance of Chosen Portfolio", linestyle="-", linewidth=3, color=colors[5])
plt.axhspan(100*np.mean(mu[sorted_indices][-K:]), 100*np.mean(means[-K:]), alpha=1, color='lightgray', zorder=0, label="Post-Decision Disappointment")

# Add labels and title
plt.xlabel('Projects')
plt.ylabel('Return on Investments [\%]')
plt.legend(fontsize=def_font_size-4)

# Save figure
plt.savefig('./PlotsTheory/OptimizersCurse2.pdf', dpi=150)

# Show the plot
plt.show()
```
